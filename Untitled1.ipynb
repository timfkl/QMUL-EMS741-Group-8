```python
import os
import numpy as np
from glob import glob
from PIL import Image
import matplotlib.pyplot as plt

import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from tqdm import tqdm
```

```python
import torchvision.transforms as T

transform_img = T.Compose([
    T.Resize((256, 256)),
    T.ToTensor()
])

transform_mask = T.Compose([
    T.Resize((256, 256)),
    T.ToTensor()
])


class SegmentationTaskDataset(Dataset):
    def __init__(self, image_paths, mask_paths):
        self.image_paths = image_paths
        self.mask_paths = mask_paths

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        img = Image.open(self.image_paths[idx]).convert("L")
        mask = Image.open(self.mask_paths[idx]).convert("L")

        img = transform_img(img)
        mask = transform_mask(mask)
        mask = (mask > 0.5).float()

        return img, mask


def load_tasks(root):
    tasks = {}
    for task_dir in sorted(os.listdir(root)):
        img_dir = os.path.join(root, task_dir, "images")
        mask_dir = os.path.join(root, task_dir, "masks")

        images = sorted(glob(img_dir + "/*.png"))
        masks = sorted(glob(mask_dir + "/*.png"))

        tasks[task_dir] = SegmentationTaskDataset(images, masks)

    return tasks
```

```python
train_tasks = load_tasks("./train")
val_tasks   = load_tasks("./val")
test_tasks  = load_tasks("./test")

train_tasks.keys(), val_tasks.keys(), test_tasks.keys()
```

```python
class DoubleConv(nn.Module):
    def __init__(self, in_ch, out_ch):
        super().__init__()
        self.net = nn.Sequential(
            nn.Conv2d(in_ch, out_ch, 3, padding=1),
            nn.ReLU(),
            nn.Conv2d(out_ch, out_ch, 3, padding=1),
            nn.ReLU()
        )

    def forward(self, x):
        return self.net(x)


class UNet(nn.Module):
    def __init__(self):
        super().__init__()
        self.down1 = DoubleConv(1, 32)
        self.down2 = DoubleConv(32, 64)
        self.pool = nn.MaxPool2d(2)

        self.up = nn.ConvTranspose2d(64, 32, 2, stride=2)
        self.final = nn.Conv2d(32, 1, 1)

    def forward(self, x):
        x1 = self.down1(x)
        x2 = self.down2(self.pool(x1))
        x = self.up(x2)
        x = self.final(x)
        return torch.sigmoid(x)
```

```python
def reptile_step(model, task_loader, inner_steps=5, lr_inner=1e-3, meta_lr=1e-2):
    temp = UNet()
    temp.load_state_dict(model.state_dict())

    opt = optim.Adam(temp.parameters(), lr=lr_inner)

    for _ in range(inner_steps):
        img, mask = next(iter(task_loader))
        pred = temp(img)
        loss = nn.BCELoss()(pred, mask)
        opt.zero_grad()
        loss.backward()
        opt.step()

    meta_state = model.state_dict()
    temp_state = temp.state_dict()

    for k in meta_state:
        meta_state[k] += meta_lr * (temp_state[k] - meta_state[k])

    model.load_state_dict(meta_state)
```

```python
def meta_train(model, tasks, epochs=200):
    task_names = list(tasks.keys())

    for epoch in range(epochs):
        task = np.random.choice(task_names)
        dataset = tasks[task]
        loader = DataLoader(dataset, batch_size=4, shuffle=True)

        reptile_step(model, loader)

        if epoch % 20 == 0:
            print(f"Epoch {epoch} completed.")
```

```python
def adapt(model, dataset, n_shot=1, steps=20, lr=1e-3):
    adapted = UNet()
    adapted.load_state_dict(model.state_dict())

    opt = optim.Adam(adapted.parameters(), lr=lr)

    few = torch.utils.data.Subset(dataset, list(range(n_shot)))
    loader = DataLoader(few, batch_size=1, shuffle=True)

    for _ in range(steps):
        img, mask = next(iter(loader))
        pred = adapted(img)
        loss = nn.BCELoss()(pred, mask)
        opt.zero_grad()
        loss.backward()
        opt.step()

    return adapted
```

```python
def dice(pred, target, eps=1e-6):
    pred = (pred > 0.5).float()
    inter = (pred * target).sum()
    union = pred.sum() + target.sum()
    return (2 * inter + eps) / (union + eps)
```

```python
def evaluate(model, dataset):
    loader = DataLoader(dataset, batch_size=1)
    scores = []

    for img, mask in loader:
        pred = model(img)
        scores.append(dice(pred, mask).item())

    return np.mean(scores)
```

```python
def show_prediction(model, dataset, idx=0):
    img, mask = dataset[idx]
    with torch.no_grad():
        pred = model(img.unsqueeze(0))[0]

    pred_bin = (pred > 0.5).float()

    fig, ax = plt.subplots(1, 3, figsize=(12, 4))
    ax[0].imshow(img.squeeze(), cmap="gray")
    ax[0].set_title("Input MRI")

    ax[1].imshow(mask.squeeze(), cmap="gray")
    ax[1].set_title("Ground Truth")

    ax[2].imshow(pred_bin.squeeze(), cmap="gray")
    ax[2].set_title("Prediction")

    for a in ax:
        a.axis("off")

    plt.show()
```

```python
model = UNet()
meta_train(model, train_tasks, epochs=200)
```

```python
for task_name, dataset in test_tasks.items():
    print(f"\n=== {task_name} ===")

    for n in [1, 3, 5]:
        adapted = adapt(model, dataset, n_shot=n)
        score = evaluate(adapted, dataset)
        print(f"{n}-shot Dice: {score:.3f}")
```

```python
task_name = list(test_tasks.keys())[0]
dataset = test_tasks[task_name]

print("Before adaptation")
show_prediction(model, dataset, idx=0)

print("After 1-shot adaptation")
adapted_1 = adapt(model, dataset, n_shot=1)
show_prediction(adapted_1, dataset, idx=0)

print("After 5-shot adaptation")
adapted_5 = adapt(model, dataset, n_shot=5)
show_prediction(adapted_5, dataset, idx=0)
```
